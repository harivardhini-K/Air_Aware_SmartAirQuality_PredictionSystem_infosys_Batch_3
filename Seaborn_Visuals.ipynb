```python
#Seaborn Data Visualizations - Titanic Dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("C:\\Users\\Admin\\Downloads\\train_dataset.csv")

# Apply a clean theme
sns.set(style="whitegrid")
```

```python
#1.Age Distribution - Histogram with KDE
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=25, kde=True, color='steelblue')
plt.title('Age Distribution of Passengers')
plt.show()
```

```python
# 2.Survival Count - Countplot
plt.figure(figsize=(6,5))
sns.countplot(x='Survived', data=df, palette='pastel')
plt.title('Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
#3.Passenger Class vs Survival - Stacked Countplot
plt.figure(figsize=(7,5))
sns.countplot(x='Pclass', hue='Survived', data=df, palette='Set2')
plt.title('Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
#4.Age vs Fare - Scatterplot
plt.figure(figsize=(8,5))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df, palette='coolwarm', alpha=0.7)
plt.title('Age vs Fare (Colored by Survival)')
plt.show()
```

```python
# 5.Fare by Passenger Class - Boxplot
plt.figure(figsize=(8,5))
sns.boxplot(x='Pclass', y='Fare', data=df, palette='muted')
plt.title('Fare Distribution by Passenger Class')
plt.show()
```

```python
# 6.Age Distribution by Gender - Violin Plot
plt.figure(figsize=(8,5))
sns.violinplot(x='Sex', y='Age', data=df, palette='husl', inner='quartile')
plt.title('Age Distribution by Gender')
plt.show()
```

```python
# 7.Correlation Heatmap
plt.figure(figsize=(8,6))
numeric_cols = ['Survived', 'Age', 'Fare', 'SibSp', 'Parch', 'Pclass']
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

```

```python
# 8. Pairplot of Key Numeric Features
sns.pairplot(df[['Survived', 'Age', 'Fare', 'SibSp', 'Parch']], hue='Survived', palette='husl')
plt.suptitle('Pairplot of Titanic Numeric Features', y=1.02)
plt.show()
```

```python
#8.Survival Rate by Embarkation Port - Barplot
plt.figure(figsize=(7,5))
sns.barplot(x='Embarked', y='Survived', data=df, palette='Set3')
plt.title('Average Survival Rate by Embarkation Port')
plt.show()
```

```python
#10.Age vs Fare - Regression Plot (Trend Line)
plt.figure(figsize=(8,5))
sns.regplot(x='Age', y='Fare', data=df, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Fare vs Age with Regression Line')
plt.show()
```
