```python
# Plotly Data Visualizations - Titanic Dataset (10 Unique Plots)
!pip install plotly 
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Load dataset
df = pd.read_csv("C:\\Users\\Admin\\Downloads\\train_dataset.csv")
```

```python
import plotly.io as pio
pio.renderers.default = 'notebook_connected'
```

```python
#1.Age Distribution Histogram
fig = px.histogram(df, x='Age', nbins=25, color_discrete_sequence=['blue'], title='Age Distribution of Passengers')
fig.show()
```

```python
#2.Survival Count (Bar Chart)
# Calculate survival counts and rename columns
survival_counts = df['Survived'].value_counts().reset_index()
survival_counts.columns = ['Survived_Status', 'Count']

# Create the bar chart
import plotly.express as px
fig = px.bar(
    survival_counts,
    x='Survived_Status',
    y='Count',
    title='Survival Count',
    labels={'Survived_Status': 'Survived (0 = No, 1 = Yes)', 'Count': 'Number of Passengers'},
    color='Survived_Status',
    color_discrete_sequence=['red', 'green']
)
fig.show()
```

```python
#3.Passenger Class vs Survival (Stacked Bar)
pclass_surv = df.groupby(['Pclass', 'Survived']).size().reset_index(name='Count')
fig = px.bar(pclass_surv, x='Pclass', y='Count', color='Survived',
             title='Survival by Passenger Class', barmode='group',
             color_continuous_scale='Tealgrn')
fig.show()
```

```python
#4.Age vs Fare (Interactive Scatter Plot)
fig = px.scatter(df, x='Age', y='Fare', color='Survived', size='Fare',
                 hover_data=['Sex', 'Pclass', 'Embarked'],
                 title='Age vs Fare (Interactive Scatter)',
                 color_continuous_scale='Viridis')
fig.show()
```

```python
#5.Fare by Passenger Class (Box Plot)
fig = px.box(df, x='Pclass', y='Fare', color='Pclass',
             title='Fare Distribution by Passenger Class')
fig.show()
```

```python
#6.Age Distribution by Gender (Violin Plot)
fig = px.violin(df, x='Sex', y='Age', color='Sex', box=True, points='all',
                title='Age Distribution by Gender')
fig.show()
```

```python
# 7. Correlation Heatmap (Interactive)
corr = df[['Survived', 'Age', 'Fare', 'SibSp', 'Parch', 'Pclass']].corr()
fig = px.imshow(corr, text_auto=True, color_continuous_scale='RdBu_r', title='Correlation Heatmap')
fig.show()
```

```python
# 8.Survival Rate by Embarkation Port (Bar Plot)
embarked_survival = df.groupby('Embarked')['Survived'].mean().reset_index()
fig = px.bar(embarked_survival, x='Embarked', y='Survived',
             title='Average Survival Rate by Embarkation Port',
             color='Embarked', color_discrete_sequence=px.colors.qualitative.Set2)
fig.show()
```

```python
# 9. Average Fare by Age (Line Plot)
age_fare = df[['Age', 'Fare']].dropna().groupby('Age')['Fare'].mean().reset_index()
fig = px.line(age_fare, x='Age', y='Fare', title='Average Fare by Age', markers=True)
fig.show()
```

```python
#10.3D Scatter Plot - Age, Fare, Pclass Colored by Survival
fig = px.scatter_3d(df, x='Age', y='Fare', z='Pclass', color='Survived',
                    title='3D Scatter: Age, Fare, Class, and Survival',
                    color_continuous_scale='Plasma')
fig.show()
```
