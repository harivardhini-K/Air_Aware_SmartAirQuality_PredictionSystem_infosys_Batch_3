```python
# Matplotlib Data Visualizations - Titanic Dataset
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
df = pd.read_csv("C:\\Users\\Admin\\Downloads\\train_dataset.csv")

# Display first few rows
print(df.head())
```

```python
#!.Age Distribution Histogram
# ------------------------------------------------------------
plt.figure(figsize=(8,5))
plt.hist(df['Age'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
#2 Survival Count (Bar Chart)
plt.figure(figsize=(6,5))
df['Survived'].value_counts().plot(kind='bar', color=['red', 'green'])
plt.title('Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
#3.Gender Distribution (Pie Chart)
plt.figure(figsize=(6,6))
df['Sex'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightskyblue'])
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

```python
# 4.Age vs Fare (Scatter Plot)
plt.figure(figsize=(8,5))
plt.scatter(df['Age'], df['Fare'], alpha=0.5, c='purple')
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
# 5.Fare Distribution by Class (Box Plot)
plt.figure(figsize=(8,5))
df.boxplot(column='Fare', by='Pclass', grid=False)
plt.title('Fare Distribution by Passenger Class')
plt.suptitle('')
plt.xlabel('Class')
plt.ylabel('Fare')
plt.show()
```

```python
# 6.Age vs Fare (Hexbin Plot)
plt.figure(figsize=(8,5))
plt.hexbin(df['Age'], df['Fare'], gridsize=30, cmap='Blues')
plt.title('Age vs Fare Density (Hexbin)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.colorbar(label='Count')
plt.show()
```

```python
# 7. Average Fare by Age (Line Plot)
plt.figure(figsize=(8,5))
age_fare = df[['Age', 'Fare']].dropna().groupby('Age')['Fare'].mean()
plt.plot(age_fare.index, age_fare.values, color='darkorange')
plt.title('Average Fare by Age')
plt.xlabel('Age')
plt.ylabel('Average Fare')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
#8.Survival Rate by Passenger Class (Area Plot)
plt.figure(figsize=(8,5))
survival_rate = df.groupby('Pclass')['Survived'].mean()
plt.fill_between(survival_rate.index, survival_rate.values, color='teal', alpha=0.5)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
#9.Age Distribution by Survival (Overlaid Histograms)
plt.figure(figsize=(8,5))
plt.hist(df[df['Survived']==0]['Age'].dropna(), bins=20, alpha=0.5, label='Did Not Survive', color='red')
plt.hist(df[df['Survived']==1]['Age'].dropna(), bins=20, alpha=0.5, label='Survived', color='green')
plt.title('Age Distribution by Survival Status')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
#10.Correlation Heatmap (Matplotlib)
plt.figure(figsize=(8,6))
corr = df[['Survived', 'Age', 'Fare', 'SibSp', 'Parch', 'Pclass']].corr()
plt.imshow(corr, cmap='coolwarm', interpolation='nearest')
plt.title('Correlation Heatmap')
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.colorbar(label='Correlation Coefficient')
plt.show()
```
