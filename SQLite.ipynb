```python
import sqlite3
import pandas as pd

# Connect to your database (replace with your actual path)
conn = sqlite3.connect(r"C:\Users\Dell\Downloads\Milestone 2\SQLite Test.db")

# 1. Group by occupation
query1 = "SELECT occupation FROM dataset_1 GROUP BY occupation;"
df1 = pd.read_sql_query(query1, conn)
display(df1)


```

```python
# 2. Average temperature per weather
query2 = """
SELECT weather, AVG(temperature) AS avg_temp
FROM dataset_1
GROUP BY weather;
"""
df2 = pd.read_sql_query(query2, conn)
display(df2)
```

```python
# 3. Count of temperature per weather
query3 = """
SELECT weather, COUNT(temperature) AS count_temp
FROM dataset_1
GROUP BY weather;
"""
df3 = pd.read_sql_query(query3, conn)
display(df3)

# 4. Count of distinct temperature per weather
query4 = """
SELECT weather, COUNT(DISTINCT temperature) AS count_distinct_temp
FROM dataset_1
GROUP BY weather;
"""
df4 = pd.read_sql_query(query4, conn)
display(df4)
```

```python
# 5. Sum of temperature per weather
query5 = """
SELECT weather, SUM(temperature) AS sum_temp
FROM dataset_1
GROUP BY weather;
"""
df5 = pd.read_sql_query(query5, conn)
display(df5)

# 6. Minimum temperature per weather
query6 = """
SELECT weather, MIN(temperature) AS min_temp
FROM dataset_1
GROUP BY weather;
"""
df6 = pd.read_sql_query(query6, conn)
display(df6)

# 7. Maximum temperature per weather
query7 = """
SELECT weather, MAX(temperature) AS max_temp
FROM dataset_1
GROUP BY weather;
"""
df7 = pd.read_sql_query(query7, conn)
display(df7)
```

```python
# 8. Occupation = 'Student'
query8 = """
SELECT occupation
FROM dataset_1
GROUP BY occupation
HAVING occupation = 'Student';
"""
df8 = pd.read_sql_query(query8, conn)
display(df8)
```

```python
# 9. UNION dataset_1 and table_to_union
query9 = """
SELECT * FROM dataset_1
UNION
SELECT * FROM table_to_union;
"""
df9 = pd.read_sql_query(query9, conn)
display(df9)
```

```python
# 10. DISTINCT destination from UNION
query10 = """
SELECT DISTINCT destination
FROM (
    SELECT * FROM dataset_1
    UNION
    SELECT * FROM table_to_union
);
"""
df10 = pd.read_sql_query(query10, conn)
display(df10)
```

```python
# 11. INNER JOIN dataset_1 and table_to_join on time
query11 = """
SELECT a.destination, a.time, b.part_of_day
FROM dataset_1 a
INNER JOIN table_to_join b
ON a.time = b.time;
"""
df11 = pd.read_sql_query(query11, conn)
display(df11)
```

```python
# 12. LIKE operator
query1 = """
SELECT *
FROM dataset_1
WHERE weather LIKE 'Sun%';
"""
df1 = pd.read_sql_query(query1, conn)
display(df1)

# 13. BETWEEN operator
query2 = """
SELECT DISTINCT temperature 
FROM dataset_1
WHERE temperature BETWEEN 29 AND 75;
"""
df2 = pd.read_sql_query(query2, conn)
display(df2)

# 14. IN operator
query3 = """
SELECT occupation
FROM dataset_1
WHERE occupation IN ('Sales & Related', 'Management');
"""
df3 = pd.read_sql_query(query3, conn)
display(df3)
```

```python
# 15. Window Function - OVER and PARTITION BY
query4 = """
SELECT 
    destination, 
    weather, 
    AVG(temperature) OVER (PARTITION BY weather) AS avg_temp_by_weather
FROM dataset_1;
"""
df4 = pd.read_sql_query(query4, conn)
display(df4)

# 16. Window Function with ORDER BY
query5 = """
SELECT 
    destination, 
    weather, 
    AVG(temperature) OVER (PARTITION BY weather ORDER BY destination) AS avg_temp_by_weather
FROM dataset_1;
"""
df5 = pd.read_sql_query(query5, conn)
display(df5)


# 17. ROW_NUMBER()
query6 = """
SELECT 
    destination, 
    weather, 
    ROW_NUMBER() OVER (PARTITION BY weather ORDER BY destination) AS row_num
FROM dataset_1;
"""
df6 = pd.read_sql_query(query6, conn)
display(df6)

# 18. RANK()
query7 = """
SELECT 
    destination, 
    weather, 
    RANK() OVER (PARTITION BY weather ORDER BY destination) AS rank_num
FROM dataset_1;
"""
df7 = pd.read_sql_query(query7, conn)
display(df7)

# 19. DENSE_RANK()
query8 = """
SELECT 
    destination, 
    weather, 
    DENSE_RANK() OVER (PARTITION BY weather ORDER BY destination) AS dense_rank_num
FROM dataset_1;
"""
df8 = pd.read_sql_query(query8, conn)
display(df8)
```

```python
# 20. NTILE()
query9 = """
SELECT 
    time, 
    NTILE(50) OVER (ORDER BY time) AS ntile_bucket
FROM dataset_1;
"""
df9 = pd.read_sql_query(query9, conn)
display(df9)
```

```python
# 22. LAG()
query10 = """
SELECT 
    destination, 
    time, 
    LAG(row_count, 1, '99999') OVER (ORDER BY row_count) AS LaggedCount
FROM dataset_1;
"""
df10 = pd.read_sql_query(query10, conn)
display(df10)
```

```python
# 11. LEAD()
query11 = """
SELECT 
    destination, 
    time, 
    LEAD(row_count, 1, '99999') OVER (ORDER BY row_count) AS LeadCount
FROM dataset_1;
"""
df11 = pd.read_sql_query(query11, conn)
display(df11)

# Close connection
conn.close()
```
